# Primeira parte

>**Para esta primeira etapa, o conjunto de dados escolhido foi o wine.**

```python

wine = datasets.load_wine()
wine = pd.DataFrame(data= np.c_[wine['data'], wine['target']],
                     columns= wine['feature_names'] + ['target'])

wine
```

>## **Atividade 1**

>>Hipótese 1: O conjunto de dados wine permite classificar a qualidade de um determinado tipo de vinho baseado em suas características.

>>Hipótese 2: O teor alcoólico de um vinho indica a qualidade do mesmo. Assim, como a variação do atributo alcohol no conjunto não é muito grande, talvez este não seja um bom atributo para a classificação.

>>Hipótese 3: Atributos que possuem uma maior variação por classe, podem ser mais representativos.

>## **Atividade 2**

>>Características dos atributos

>>Alcohol: quantidade de álcool presente no vinho. 
Malic_acid: quantidade de ácido málico, define a acidez do vinho.
Color intensity: intensidade da cor do vinho
Hue: 
Proline:

>>Com base nas informações acima, escolhemos os atributos alcohol, malic_acid, proline, hue e color_intensity. Todos são constituidos de atributos contínuos e achamos, de acordo com pesquisas direcionadas em cada atributo, que os mesmos influenciam na classificação da qualidade de um determinado tipo de vinho.

```python
mean = {'malic_acid_mean_per_class': wine['malic_acid'].groupby(by=[wine.target]).mean(), 'malic_acid_mean': wine.malic_acid.mean(),
       'alcohol_mean_per_class': wine['alcohol'].groupby(by=[wine.target]).mean(), 'alcohol_mean': wine.alcohol.mean(),
       'proline_acid_mean_per_class': wine['proline'].groupby(by=[wine.target]).mean(), 'proline_mean': wine.proline.mean(),
       'hue_acid_mean_per_class': wine['hue'].groupby(by=[wine.target]).mean(), 'hue_mean': wine.hue.mean(),
       'color_intensity_mean_per_class': wine['color_intensity'].groupby(by=[wine.target]).mean(), 'color_intensity_mean': wine.color_intensity.mean()}

standard_deviation = {'malic_acid_std_per_class': wine['malic_acid'].groupby(by=[wine.target]).std(), 'malic_acid_std': wine.malic_acid.std(),
       'alcohol_std_per_class': wine['alcohol'].groupby(by=[wine.target]).std(), 'alcohol_std': wine.alcohol.std(),
       'proline_acid_std_per_class': wine['proline'].groupby(by=[wine.target]).std(), 'proline_std': wine.proline.std(),
       'hue_acid_std_per_class': wine['hue'].groupby(by=[wine.target]).std(), 'hue_std': wine.hue.std(),
       'color_intensity_std_per_class': wine['color_intensity'].groupby(by=[wine.target]).std(), 'color_intensity_std': wine.color_intensity.std()}

median = {'malic_acid_median_per_class': wine['malic_acid'].groupby(by=[wine.target]).median(), 'malic_acid_median': wine.malic_acid.median(),
       'alcohol_median_per_class': wine['alcohol'].groupby(by=[wine.target]).median(), 'alcohol_median': wine.alcohol.median(),
       'proline_acid_median_per_class': wine['proline'].groupby(by=[wine.target]).median(), 'proline_median': wine.proline.median(),
       'hue_acid_median_per_class': wine['hue'].groupby(by=[wine.target]).median(), 'hue_median': wine.hue.median(),
       'color_intensity_median_per_class': wine['color_intensity'].groupby(by=[wine.target]).median(), 'color_intensity_median': wine.color_intensity.median()}

means = pd.DataFrame(mean)
standard_deviations = pd.DataFrame(standard_deviation)
medians = pd.DataFrame(median)

means
```

```python
standard_deviations
```

```python
medians
```

>## **Atividade 3**

>Podemos observar, pelas medidas computadas acima, algumas peculiaridades sobre algumas variáveis. 

>>Por exemplo, a média do atributo álcool, bem como sua média por classe, não varia muito, o que sugere que talvez este atributo não seja tão relevante para este tipo de problema de classificação. Por outro lado, vemos que o atributo que define a intensidade da cor ja é algo mais interessante, pois a classe 1 apresenta média de classe baixa em relação a média geral, enquanto que a classe 2 apresenta valor elevado com relação à geral, sendo que a classe 0 está próximo a média total, caracterizando uma variável significativa para a classificação deste. Quanto a variável hue, vê-se que a classe 2 oferece uma média baixa em relação a geral, podendo ser útil para separá-la, no entanto, a classe 0 e 1 possuem praticamente a mesma média e, assim, somente esta variável não seria necessária para poder classificar um tipo de vinho. O mesmo fato ocorre com o ácido málico, que é mais representativo na classe 2. Por fim, o atributo proline indica um bom separador para a classe 0, separando bem a mesma da 1 e 2 e, talvez, seria um bom complemento para os atributos hue e/ou ácido málico, que separam bem apenas a classe 2.

>> Para o desvio padrão, não foram observadas uma dispersão muito grande por parte das variáveis, não indicando alguma característica significativa.

>> Quanto às medianas, os valores da intensidade da cor sugerem que a separação por classes mostrou-se bem evidente e, para os atributos hue e malic_acid, as medidas se mostraram iguais a média. Capazes de separar bem a classe 2, porém mostrando valores próximos para as classes 0 e 1, que poderiam ser separadas pelo atributo proline.

>Durante todas as etapas anteriores, foram verificados 2 problemas em relação ao conjunto de dados:
>>1 - Como o conjunto de dados foi criado no Brasil, usamos o caracter '.' para separar 3 dígitos em um número, porém, o Pandas é baseado em conjuntos de dados no padrão Norte-Americano, em que este caracter indica que este número é decimal. No conjunto de dados, tivemos que inicializar o parâmetro "thousands" como '.', para que os números fossem interpretados corretamente.

>>2 - Dados faltantes. Como citado anteriormente, nas colunas NUM_SALAS_EXISTENTE e NUM_FUNCIONARIOS, foram encontrados valores null. Uma maneira correta de encontrar um valor adequado, poderia ser substitui-lo pela moda desse atributo em relação a cidade desta escola, porém, como não havia alunos matriculados não faria sentido encontrar um valor para estes atributos, pois geraria uma alteração nos resultados gerados, e assim, decidimos excluir estes exemplos.
