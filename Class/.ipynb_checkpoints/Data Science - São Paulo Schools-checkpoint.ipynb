```python
# Pandas: library that contains the main functions to manipulate de dataset
import pandas as pd
# Numpy: 
import numpy as np
# Matplotlib and Seaborn: functions to plot the results
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Some parameters used
# sep=';' - The character the separate the columns
# header=0 - line that represents the header of the dataset
# enconding='latin1' - econding that contain som sybols used in the portuguese language
# skiprows=11 - the first 11 lines contains the datatse title and the attributes descriptions, that we are going to skip
# thousand='.' - character used to separate the thousands
escolas = pd.read_csv('CADASTRO_MATRICULAS_REGIAO_SUDESTE_SP_2012.csv', sep=';', header=0, encoding='latin1', skiprows=11, thousands=".")
# The last 2 lines do not have anything
escolas = escolas.iloc[0:escolas.shape[0]-2,:]

escolas.head()
```

># Part 1 - Select some hyphoteses related to the dataset and choose the best columns that can anwser these. 

> Hyphotese 1 - The relation students_number/classroms_number is higher in public schools

> Hyphotese 2 - The relation employees_number/students_number is higher in public schools

> Hyphotese 3 - The relation computer_number/students_number is higher in public schools

>#### The collumns that can be useful to anwser these hypoteses:

>> MATTOTAL: Total number of students in a school

>> REDE: Privada if the schools is private or Pública if the school is public

>> NUM_SALAS_EXISTENTES: Number of classroms in a school

>> NUM_FUNCIONARIOS: Number of employees in a school

>> NUM_COMPUTADORES: Number of computers in a school

> These hyphoteses can define a private and a public school, so, we can use these new features to classify a school 


```python
escolas = escolas[["MATTOTAL", "REDE", "NUM_SALAS_EXISTENTES", "NUM_FUNCIONARIOS", "NUM_COMPUTADORES"]]

escolas
```

```python
# We can verify the number of Null values in these columns
escolas.info()
```

```python
# Verify all the rows that have at least one NaN/Null value
escolas[escolas.isnull().any(axis=1)]
```

The table below indicates that all the schools that have Nan/Null values don't have any students or computers, indicating that those schools are not working or even not physically exist , so, we can discart these rows.

```python
escolas = escolas.dropna()
escolas.reset_index(drop=True, inplace=True)
escolas
```

```python
new_features = pd.DataFrame()

new_features["Students/Classroms"] = np.where(escolas["NUM_SALAS_EXISTENTES"] < 1, escolas["NUM_SALAS_EXISTENTES"], escolas['MATTOTAL'].divide(escolas["NUM_SALAS_EXISTENTES"]))
new_features["Students/Computers"] = np.where(escolas["NUM_COMPUTADORES"] < 1, escolas["NUM_COMPUTADORES"], escolas['MATTOTAL'].divide(escolas["NUM_COMPUTADORES"]))
new_features["Students/Employees"] = np.where(escolas["NUM_FUNCIONARIOS"] < 1, escolas["NUM_FUNCIONARIOS"], escolas['MATTOTAL'].divide(escolas["NUM_FUNCIONARIOS"]))
new_features["REDE"] = escolas["REDE"]

new_features
```

```python
new_features.groupby(by=["REDE"]).mean()
```
